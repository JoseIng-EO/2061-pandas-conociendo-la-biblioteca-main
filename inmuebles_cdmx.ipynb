# **Demandas de Machine learning**

### ¿Cuales son los valores promedios de alquiler por cada propiedad?

```python
datos.head()
```

```python
datos['Valor'].mean()
```

```python
datos.groupby('Tipo').mean(numeric_only=True)
```

```python
datos.groupby('Tipo')[['Valor']].mean().sort_values('Valor')
```

```python
df_tipo_precio = datos.groupby('Tipo')[['Valor']].mean().sort_values('Valor')
```

```python
import matplotlib.pyplot as plt
```

```python
df_tipo_precio.plot(kind='barh',figsize=(12,8),color='purple')
plt.title('Valor Promedio de Alquiler')
plt.ylabel('Tipo de Inmueble')
```

```python
datos.Tipo.unique()
```

```python
inmuebles_comerciales = ['Conjunto Comercial/Sala','Edificio Completo','Tienda/Salón', 'Casa Comercial','Terreno Estándar',
                      'Cochera/Estacionamiento','Galpón/Depósito/Almacén','Tienda en Centro Comercial', 'Hotel',
                      'Loteo/Condominio','Industria']
```

```python
datos.query('Tipo in @inmuebles_comerciales')
```

```python
df = datos.query('Tipo not in @inmuebles_comerciales')
df
```

```python
df_tipo_precio = df.groupby('Tipo')[['Valor']].mean().sort_values('Valor')
df_tipo_precio.plot(kind='barh',figsize=(12,8),color='purple')
plt.title('Valor Promedio de Alquiler')
plt.ylabel('Tipo de Inmueble')
```

```python
df.Tipo.unique()
```

```python
porcentual = df.Tipo.value_counts(normalize=True)*100
porcentual
```

```python
porcentaje_tipo = porcentual.to_frame()
porcentaje_tipo.plot(kind='bar',figsize=(12,8),color='green',xlabel='Tipos',ylabel='Porcentaje')
plt.title('Porcentaje de Inmuebles Analizados')
```
